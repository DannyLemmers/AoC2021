```python
import numpy as np
import time

with open('fishInitial.txt') as f:
    fishState = (f.readlines())
    fishState = fishState[0]
    fishState = np.fromstring(fishState, dtype=np.int, sep=',' )
```

```python
groupedFish = np.zeros(9, dtype = np.uint64)
for j in range(len(fishState)):
    if fishState[j] == 0:
        groupedFish[0] = groupedFish[0] + 1
    elif fishState[j] == 1:
        groupedFish[1] = groupedFish[1] + 1
    elif fishState[j] == 2:
        groupedFish[2] = groupedFish[2] + 1
    elif fishState[j] == 3:
        groupedFish[3] = groupedFish[3] + 1
    elif fishState[j] == 4:
        groupedFish[4] = groupedFish[4] + 1
    elif fishState[j] == 5: 
        groupedFish[5] = groupedFish[5] + 1
    elif fishState[j] == 6: 
        groupedFish[6] = groupedFish[6] + 1
    else: 
        print('We started with a too high number. Probably forgot to reload the dataset!')
groupedFish =groupedFish[::-1]
```

```python
tend = 256
t = time.time()
for i in range(tend):
    
    groupedFishTemp = np.array(groupedFish, copy=True)  
    for j in range(len(groupedFish)):
        if j == 0:
            groupedFish[j] = groupedFishTemp[8]
        elif j == 2:
            groupedFish[j] = groupedFishTemp[8]+groupedFishTemp[j-1]
        else:
            groupedFish[j] = groupedFishTemp[j-1]

print("Total Time: " + str(time.time()-t) + 's')
print(sum(groupedFish))
```
